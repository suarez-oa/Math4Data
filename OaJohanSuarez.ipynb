 **CREO LA FUNCION SENTIMENT CON LOS VALORES ADECUADO**

```python
import nltk #importamos la libreria nltk
nltk.download('vader_lexicon') #descargamos el lexicon de vader
from nltk.sentiment.vader import SentimentIntensityAnalyzer #importamos el analizador de sentimientos
sia = SentimentIntensityAnalyzer() #creamos una variable con el analizador de sentimientos

def get_sentiment_score(text):
  if pd.isnull(text) or text == '': #si el texto es nulo o vacio devolvemos 1
    return 1
  elif isinstance(text, str):
    sentiment = sia.polarity_scores(text) #si el texto no es nulo o vacio calculamos el sentimiento
    compound_score = sentiment['compound'] #obtenemos el valor del sentimiento
    if compound_score >= -0.05: #si el valor del sentimiento es mayor o igual a -0.05 devolvemos 0
      return 0 #si el valor del sentimiento es mayor o igual a -0.05 devolvemos 0
    elif compound_score <= -0.05: #si el valor del sentimiento es menor o igual a -0.05 devolvemos 2
      return 2 #si el valor del sentimiento es menor o igual a -0.05 devolvemos 2
    else:
      return 1 #si no se cumple ninguna de las condiciones anteriores devolvemos 1
  else:
    return 1 #si no se cumple ninguna de las condiciones anteriores devolvemos 1

```

```python
muestra = df.head(10000)
tfidf = TfidfVectorizer(stop_words='english')
muestra = muestra.fillna('')

tfidf_matri = tfidf.fit_transform(muestra['review'])
cosine_similarity = linear_kernel(tfidf_matri, tfidf_matri)
```

```python
@app.get('/recomendacion_usuario/{id_juego}')
def recomendacion_juego(id_juego: int):
  if id_juego not in muestra['id'].values:
    return {'mensaje': 'No existe el id del juego.'}
  titulo = muestra.loc[muestra['id']== id_juego, 'title'].iloc[0]
  idx = muestra[muestra['title'] == titulo].index[0]
  sim_cosine = list(enumerate(cosine_similarity[idx]))
  sim_scores = sorted(sim_cosine, key=lambda x: x[1], reverse=True)
  sim_ind = [i for i, _ in sim_scores[1:6]]
  sim_juegos = muestra['title'].iloc[sim_ind].values.tolist()
  return {'juegos recomendados': list(sim_juegos)}


@app.get('/recomendacion_juego/{id_producto}')
def recomendacion_usuario(id_usuario: object):
  if id_usuario not in muestra['steam_id'].values:
    return {'mensaje': 'No existe el id del usuario.'}
  generos = muestra.columns[2:17]
  filter_df = muestra[(muestra[generos].any(axis=1)) & (muestra['steam_id'] != id_usuario)]
  tdfid_matrix_filtered = tfidf.transform(filter_df['review'])
  cosine_similarity_filtered = linear_kernel(tdfid_matrix_filtered, tdfid_matrix_filtered)
  idx = muestra[muestra['steam_id'] == id_usuario].index[0]
  sim_cosine = list(enumerate(cosine_similarity_filtered[idx]))
  sim_scores = sorted(sim_cosine, key=lambda x: x[1], reverse=True)
  sim_ind = [i for i, _ in sim_scores[1:6]]
  sim_juegos = filter_df['title'].iloc[sim_ind].values.tolist()

  return {'juegos recomendados por el usuario': list(sim_juegos)}
```

```python
def recomendacion_usuario(id_usuario: str):
  if id_usuario not in muestra['steam_id'].values:  #si el id del usuario no esta en la muestra
    return {'mensaje': 'No existe el id del usuario.'} #devuelve el mensaje
  generos = muestra.columns[2:17] #creo una lista con los generos
  filter_df = muestra[(muestra[generos].any(axis=1)) & (muestra['steam_id'] != id_usuario)] #creo un dataframe con los generos y el id del usuario
  tdfid_matrix_filtered = tfidf.transform(filter_df['review']) #aplico la matriz tfidf
  cosine_similarity_filtered = linear_kernel(tdfid_matrix_filtered, tdfid_matrix_filtered) #aplico la similitud coseno
  idx = muestra[muestra['steam_id'] == id_usuario].index[0] #devuelve el indice del usuario
  sim_cosine = list(enumerate(cosine_similarity_filtered[idx])) #devuelve la similitud coseno del indice
  sim_scores = sorted(sim_cosine, key=lambda x: x[1], reverse=True) #devuelve la similitud coseno ordenada de mayor a menor
  sim_ind = [i for i, _ in sim_scores[1:6]] #devuelve los 5 primeros juegos
  sim_juegos = filter_df['title'].iloc[sim_ind].values.tolist() #devuelve los 5 primeros juegos en una lista

  return {'juegos recomendados por el usuario': list(sim_juegos)} #devuelve los 5 primeros juegos en una lista
```
